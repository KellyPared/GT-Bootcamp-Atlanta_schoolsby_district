```python
atl_zip_shp.rename(columns = {'ZIP':'zipcode'}, inplace = True)
atl_zip_shp
```

```python
data = pd.merge(atl_zip_shp, crime_zip, how= "left", on = ["zipcode","zipcode"])
```

```python
data
```

```python
data_new= data.drop('last_edite', axis=1)
data_new

data = data_new.dropna(how="any")
data = data.reset_index()
```

```python
import matplotlib.pyplot as plt
```

```python
#atlanta area : data.plot(column="total_crime",cmap='OrRd', legend=True, missing_kwds={'color': 'lightgrey'})
data.plot(column="total_crime",cmap='OrRd', legend=True)

data['coords'] = data['geometry'].apply(lambda x: x.representative_point().coords[:])
data['coords'] = [coords[0] for coords in data['coords']]

for i in range(0, data.shape[0]):
    plt.annotate(data["zipcode"][i], xy=(data['coords'][i][0],data['coords'][i][1]), horizontalalignment='center')
```
